## Preprocessing

```python
for image_batch, labels_batch in train_df:
    print(image_batch.shape)
    print(labels_batch.shape)
    break
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_df = train_df.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
valid_df = valid_df.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
TRAINING = "train"
TEST = "test"
VALIDATION = "valid"

training_datagen = ImageDataGenerator(
    rescale = 1./255,
    horizontal_flip = True,
)

test = ImageDataGenerator(
    rescale = 1./255
)

print ('Train Generator: ', end = "")
train_generator = training_datagen.flow_from_directory(
    TRAINING,
    target_size = (img_h,img_w),
    shuffle = True,
    class_mode = 'categorical',
    batch_size = 32
)

print ('Valid Generator: ', end = "")
valid_generator = test.flow_from_directory(
    VALIDATION,
    target_size = (img_h,img_w),
    shuffle = True,
    class_mode = 'categorical',
    batch_size = 32
)

print ('Test Generator: ', end = "")
test_generator = test.flow_from_directory(
    TEST,
    target_size = (img_h,img_w),
    shuffle = True,
    class_mode = 'categorical',
    batch_size = 32
)
```

## Data Modeling

```python
num_classes = len(class_name)

model =Sequential([
      
      layers.Conv2D(64, 3, padding='same', activation='relu', input_shape = (img_h, img_w, 3)),
      layers.MaxPooling2D(),
      layers.Conv2D(32, 3, padding='same', activation='relu'),
      layers.MaxPooling2D(),
      layers.Conv2D(16, 3, padding='same', activation='relu'),
      layers.MaxPooling2D(),
      layers.Flatten(),
      layers.Dense(128, activation='relu'),
      layers.Dense(128, activation='relu'),
      layers.Dense(num_classes, activation='softmax')
])
```

```python
model.summary()
```

## Training Model

```python
model.compile(optimizer="Adamax", loss="categorical_crossentropy", metrics=['accuracy'])
```

```python
epochs = 200
history2 = model.fit(
    train_generator,
    validation_data=valid_generator,
    steps_per_epoch= 15,
    epochs=epochs
)
```

```python
model.save_weights('model.h5')
```

## Evaluasi dan Test Model

```python
#train
y_acc_train = history2.history['accuracy']
y_acc_val = history2.history['val_accuracy']

#validation
y_loss_train = history2.history['loss']
y_loss_val = history2.history['val_loss']

#epoch
x_ = range(len(y_acc_train))
```

```python
plt.figure(1, figsize=(5,4))
plt.plot(x_, y_acc_train, c='r', label='Training')
plt.plot(x_, y_acc_val, c='b', label='Validation')
plt.title('Akurasi')
plt.xlabel('Epochs')
plt.ylabel('Akurasi')
plt.legend()

plt.figure(2, figsize=(5,4))
plt.plot(x_[3:], y_loss_train[3:], c='r', label='Training')
plt.plot(x_[3:], y_loss_val[3:], c='b', label='Validation')
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
train_acc = model.evaluate(train_generator)
valid_acc = model.evaluate(valid_generator)
test_acc  = model.evaluate(test_generator)

print(train_acc)
print(valid_acc)
print(test_acc)
```
